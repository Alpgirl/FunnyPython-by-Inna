## NUMBER-SQUARES

```python
class Iterator:
    def __iter__(self):
        return self
    def __init__(self, lim):
        self.limit = lim
        self.counter = 0
    def __next__(self):
        if self.counter < self.limit:
            self.counter += 1
            return self.counter**2
        else:
            raise StopIteration
```

```python
def generate_Iterator(n_max):
    for el in range(1, n_max + 1):
        yield el**2

def generate_expression(n_max):
    return [el**2 for el in range(1, n_max + 1)]
        
        
N = int(input('N = '))

iter1 = Iterator(N)
print('Класс-итератор:')
for i in iter1:
    print(i, end=' ')

print('\nФункция-генератор:')
for k in generate_Iterator(N):
    print(k, end=' ')

print('\nГенераторное выражение:')
for l in generate_expression(N):
    print(l, end=' ')
```

## REFACTORING

```python
def generator(lst1, lst2):
    for x in lst1:
        for y in lst2:
            yield x, y

            
list_1 = [2, 5, 7, 10]
list_2 = [3, 8, 4, 9]
to_find = 56
for x, y in generator(list_1, list_2):
    result = x * y
    print(x, y, result)
    if x * y == 56: 
        print('Found!!!')
        break
```

## FILE-PATHS

```python
import os

def gen_files_path(el):
    path = el[0]
    for el2 in el[1:]:
        if isinstance(el2, list):
            for el3 in el2:
                path_item = path + '/' + el3
                if os.path.isfile(path_item):
                    print(path_item)
                    
                    
name_directory = input('Введите название каталога: ')
print(os.walk(name_directory))

for el in os.walk('../'+name_directory):
    gen_files_path(el)
    print()
```

## SEQUENCE-HOFSHTADTERA

Фигура Хофштадтера

R(1) = 1 ; S(1) = 2 
R(n) = R(n-1)+ S(n-1),  n > 1

R: 1, 3, 7, 12, 18, 26, 35, 45, 56, 69, 83, 98, 114, 131, 150, 170, 191, 213, 236, 260, ... ( последовательность A005228 в OEIS)

S: 2, 4, 5, 6, 8, 9, 10, 11, 13, 14, 15, 16, 17, 19, 20, 21, 22, 23, 24, 25, ... ( последовательность A030124 в OEIS)

Последовательность Хофштадтера Q

Q(1) = 1, Q(2) = 2
Q(n)=Q(n−Q(n−1))+Q(n−Q(n−2))

Q: 1, 1, 2, 3, 3, 4, 5, 5, 6, 6, 6, 8, 8, 8, 10, 9, 10, 11, 11, 12, ... 

```python
def Q(n):
    if n < 3:
        return 1
    return Q(n-Q(n-1)) + Q(n-Q(n-2))

def sequence(lim):
    for i in range(1, lim):
        yield Q(i)
        
        
for el in sequence(20):
    print(el, end=' ')
```

## ROWS-AMOUNT

```python
import os
import pathlib

def read_file(name_directory, path):
    s = 0
    for el in os.listdir(name_directory):
        print(path + el)
        if os.path.isfile(path + el) and el != '.DS_Store':
            for line in open(path + el, "r"):
                if line != '\n' and not line.startswith('#'):
                    s += 1
            yield s
            

name_directory = input('Введите название каталога с питоновскими файлами: ')
#path = str(pathlib.Path().resolve()).split('/')
#path = path[len(path)-1] + '/' + name_directory + '/'
path = name_directory + '/'
for el in read_file(name_directory, path):
    print(el)
```

## SINGLY-LINKED-LIST

```python
class Node:
    def __init__(self, dataval = None):
        self.dataval = dataval
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None
        
    def append(self, newdata):
        NewNode = Node(newdata)
        
        if self.headval is None:
            self.headval = NewNode
            return
        
        laste = self.headval
        while laste.nextval is not None:
            laste = laste.nextval
        laste.nextval = NewNode

    def remove(self, ind):
        to_save = ind
        HeadVal = self.headval
        
        # remove head element if ind = 0 
        if HeadVal is not None:
            if ind == 0:
                self.headval = HeadVal.nextval
                HeadVal = None
                return
            
        # find previous element by ind   
        while HeadVal is not None:
            if ind == 0:
                break
            prev = HeadVal
            HeadVal = HeadVal.nextval
            ind -= 1
            
        # if linkedlst is empty, just return
        if HeadVal == None:
            return
        
        # skip element
        prev.nextval = HeadVal.nextval
        HeadVal = None
        print(f'Удаление {to_save + 1}-го элемента.')
        
    def get(self, ind):
        to_save = ind
        HeadVal = self.headval
        if HeadVal is None:
            return
        while HeadVal is not None:
            if ind == 0:
                print(f'Получение {to_save + 1}-го элемента:', end=' ')
                return HeadVal.dataval
            HeadVal = HeadVal.nextval
            ind -= 1
            
            
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval, end=' ')
            printval = printval.nextval
        print()
```

```python
my_list = SLinkedList()
my_list.headval = Node(0)

my_list.append(10)
my_list.append(20)
my_list.append(30)
my_list.append(60)
my_list.append(39)
my_list.append(48)
print('Текущий список: ')
my_list.listprint()
my_list.remove(3)
print('Текущий список: ')
my_list.listprint()
print(my_list.get(5))
```
