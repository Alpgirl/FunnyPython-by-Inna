## TAXES

```python
class Property:
    def __init__(self, worth):
        self.worth = worth
        
class Appartment(Property):
    def __init__(self, worth):
        super().__init__(worth)
        self.tax = self.worth/1000
class Car(Property):
    def __init__(self, worth):
        super().__init__(worth)
        self.tax = self.worth/200
class CountryHouse(Property):
    def __init__(self, worth):
        super().__init__(worth)
        self.tax = self.worth/500
```

```python
worth_prop, property1 = [0] * 3, [0] * 3
print('Привет')
money = int(input('Сколько у вас денег? '))
print('Сколько стоит ваше имущество? Если имущество отсутствует, введите 0.')
worth_prop[0] = int(input('Квартира : '))
worth_prop[1] = int(input('Машина : '))
worth_prop[2] = int(input('Дом : '))

property1[0] = Appartment(worth_prop[0])
property1[1] = Car(worth_prop[1])
property1[2] = CountryHouse(worth_prop[2]) 

s_tax = 0
for i in range(3):
    s_tax += property1[i].tax

if money < s_tax: print('Не хватаает:', s_tax - money)
else: print('Все уплачено.')
```

## OWN-DICTIONARY

```python
class MyDict:
    def __init__(self, d):
        self.dict = d
    def check(self, key):
        if self.dict.get(key) == None: return 0
        else: return self.dict.get(key)
```

```python
test_dict = {
             1: 'a',
             2 : 'b',
             3 : 'c', 
             4 : 'd' }

mydict1 = MyDict(test_dict)
print(mydict1.check(6))
```

## COMPANY

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
class Employee(Person):
    def __init__(self, name, age, salary):
        self.salary = salary
        super().__init__(name, age)
        
class Manager(Employee):
    def __init__(self, name, age, salary):
        super().__init__(name, age, salary)
        
class Agent(Employee):
    def __init__(self, name, age, sales):
        self.sales = sales
        super().__init__(name, age, 5000 + 0.05 * self.sales)
        
class Worker(Employee):
    def __init__(self, name, age, hours):
        self.work_hours = hours
        super().__init__(name, age, 100 * self.work_hours)
```

```python
import random

managers = [Manager(f'Менеджер {i}', random.randint(25,40), 13000) for i in range(1, 4)]
agents = [Agent(f'Агент {i}', random.randint(25,40), random.randint(100,1000)) for i in range(1, 4)]
workers = [Worker(f'Работник {i}', random.randint(25,40), random.randint(5,80)) for i in range(1, 4)]

for i in range(3):
    print(managers[i].name, ':', managers[i].salary, ',', agents[i].name, ':', agents[i].salary, ',', workers[i].name, ':', workers[i].salary)
```

## CAR

```python
import math

class car:
    def __init__(self, x, y, alpha):
        self.x_coord = x
        self.y_coord = y
        self.angle = math.radians(alpha)
    def move(self, distance):
        self.x_coord += distance * math.cos(self.angle)
        self.y_coord += distance * math.sin(self.angle)
    def turn(self, betta):
        self.angle += math.radians(betta)
        
class bus(car):
    def __init__(self, x, y, alpha, n, money):
        super().__init__(x, y, alpha)
        self.passengers = n
        self.money = money
    def come_in(self, n1):
        self.passengers += n1
        self.money += n1 * 55
    def get_off(self, n2):
        self.passengers += n2
```

```python
car1 = car(2, 5, 30)
bus1 = bus(0, 3, 45, 5, 0)

bus1.come_in(3)
bus1.turn(-45)
bus1.move(3)
print(bus1.money, bus1.passengers, bus1.angle, bus1.x_coord, bus1.y_coord)
```

## STEK

```python
class Stack:
    def __init__(self):
        self.elements = []
    def add(self, elem):
        self.elements.append(elem)
    def pop(self):
        if len(self.elements) > 0: return self.elements.pop()
        else: print("empty stack")


class TaskManager:
    def __init__(self):
        self.manager = dict()       
            
    def new_task(self, name, priority):
        if priority in self.manager.keys():
            if name not in self.manager[priority]:
                self.manager[priority].append(name)
        else:
            self.manager[priority] = [name]
        
    def print_tasks(self):
        self.manager = dict(sorted(self.manager.items()))
        for i in self.manager.keys():
            print(i, self.manager[i])
    
    def delete_task(self, name):
        tmp = tuple(self.manager.keys())
        i = 0
        check = 0
        while (i < len(tmp) and check == 0):
            if name in self.manager[tmp[i]]:
                if len(self.manager[tmp[i]]) == 1:
                    self.manager.pop(tmp[i])
                else:
                    self.manager[tmp[i]].remove(name)
                check = 1
            i += 1
        if check == 0: print("not found")
            
```
